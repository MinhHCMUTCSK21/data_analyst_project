```python
df_final = pd.read_csv('world_cup.csv')
df_final.drop_duplicates(inplace=True)

df_final.sort_values('year', inplace=True)
```

```python
df_final
```

```python
df_final.drop(index=37, inplace=True)

```

```python
df_final['score'] = df_final['score'].str.split(' ')
```

```python
df_final['score'] = df_final['score'].apply(lambda x: x[0])
df_final
```

```python
df_final['home-score'] = df_final['score'].str.split('–',expand=True)[0]
df_final['away-score'] = df_final['score'].str.split('–',expand=True)[1]
```

```python
df_final.drop(columns=['score'], inplace=True)
```

```python
df_final.rename(columns={'home':'home-team', 'away':'away-team'}, inplace=True)
```

```python
df_final['home-score'] = df_final['home-score'].astype(int)
df_final['away-score'] = df_final['away-score'].astype(int)
```

```python
df_final['total-goals'] = df_final['home-score'] + df_final['away-score']
```

```python
df_final
```

```python
df_final.to_csv('clean_world_cup_data.csv', index=False)
```
