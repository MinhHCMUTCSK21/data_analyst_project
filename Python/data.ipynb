### Project data visualization for California cities


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
df = pd.read_csv('california_cities.csv')
df.head()
```

### Draw scatter table


```python
plt.scatter(df['longd'], df['latd'], 
            c=np.log10(df['population_total']), 
            cmap='inferno', 
            s=df['area_total_km2']/2,
            linewidths=0,
            alpha=0.9
            );
plt.style.use('tableau-colorblind10')
plt.axis('equal'); # phn b u theo x v y
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Distribution of cities in California')
area_range = [50, 100, 300, 500]
plt.colorbar(label='log$_{10}$( population_total)')
plt.clim(3,7) # gii hn mu
# v ch thch 
index = 0;
color_range=['c','r','y','g']

for area in area_range:
    plt.scatter([], [],c = color_range[index] ,s=area/2, label=str(area) + ' km$^2$')
    index += 1

plt.legend(title='City Area')

plt.figure(figsize=(8,6));


```
