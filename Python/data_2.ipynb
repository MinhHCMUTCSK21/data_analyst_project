```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df_s = df.sample(frac=0.5, random_state=100) # random sample of 50% of the data
```

```python
df_s.isnull().sum() # check for missing values
```

```python
df_s.dropna(inplace=True) # drop missing values
```

```python
df_s=df_s.drop_duplicates() # check for duplicates
```

```python
df_s.describe(include='all') # descriptive statistics
```

```python
df_s.drop(columns=['educational-num'])
```

```python
df_s.drop(columns=['capital-gain', 'capital-loss'])
```

```python
df_s['age'].describe()
```

```python
df_s['age'].plot(kind='hist')
```

```python
len(df_s['age'][(df_s['age']<=48)&(df_s['age']>=17)])
```

```python
df_s['age'].between(17,48).sum()
```

```python
len(df_s[(df_s['education']=='Masters')|(df_s['education']=='Bachelors')])
```
