```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
soup = BeautifulSoup(requests.get('https://en.wikipedia.org/wiki/Premier_League').text, 'lxml')

all_tables = soup.find_all('table')

table = 123
for item in all_tables:
    cap = item.find('caption')
    if (cap) :
        if (cap.get_text(strip=True) == '''Results of the 'Big Six' during the 2010s'''):
            table = item
            break

body = table.find('tbody')
header = body.find('tr')

header = header.find_all('th')

header = [item.get_text(strip=True) for item in header]

print(header)

data = body.find_all('tr')
table = []
for item in data :
    cols = item.find_all('td')
    cols = [item.get_text(strip=True) for item in cols]
    table.append(cols)

df = pd.DataFrame(table, columns=header)

df.drop([0,13,14], inplace=True)

df
```
